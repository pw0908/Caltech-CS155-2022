```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
# Upload problem3data.txt to a 'data' folder in Google Colab, using the "Files"
# tab on the left.
# (You can also use urllib.request.urlretrieve, as in set 1)
train_file = 'data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
alpha = range(1,10000,10)
w_ridge = np.zeros((len(alpha),9))
for i in range(len(alpha)):
    ridge = Ridge(alpha = alpha[i])
    ridge.fit(X_train, y_train)
    w_ridge[i,:] = ridge.coef_
```

```python
w_ridge[-1,:]
```

```python
plt.semilogx(alpha,w_ridge)
plt.ylabel("$w_i$",fontsize=18)
plt.xlabel("Î±",fontsize=18)
plt.ylim(-5,5)
plt.legend(['$w_1$','$w_2$','$w_3$','$w_4$','$w_5$','$w_6$','$w_7$','$w_8$','$w_9$'],loc = 'best',ncol=2, fontsize = 8,frameon=False)
plt.savefig("ridge_3.pdf")
```

```python
alpha = np.linspace(0.01,3,30)
w_lasso = np.zeros((len(alpha),9))
for i in range(len(alpha)):
    lasso = Lasso(alpha = alpha[i])
    lasso.fit(X_train, y_train)
    w_lasso[i,:] = lasso.coef_
```

```python
plt.plot(alpha,w_lasso)
plt.ylabel("$w_i$",fontsize=18)
plt.xlabel("Î±",fontsize=18)
plt.ylim(-5,5)
plt.legend(['$w_1$','$w_2$','$w_3$','$w_4$','$w_5$','$w_6$','$w_7$','$w_8$','$w_9$'],loc = 'best',ncol=2, fontsize = 8,frameon=False)
plt.savefig("lasso_3.pdf")
```
